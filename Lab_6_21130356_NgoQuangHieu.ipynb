#Task 1.
Task 1. Compare the performance of selected classification algorithms including **Random forest**, **NaiveBayes**, and **SVM** with **FASHION** dataset based on **accuracy, precision, recall, f1** measures according to **without using selection feature** and **using selection feature**.



```python
# code

def _insertTable(t, nameAlg, y_test, y_pred):
  t.add_row([nameAlg, metrics.accuracy_score(y_test, y_pred), metrics.precision_score(y_test, y_pred, average='macro'),metrics.recall_score(y_test, y_pred, average='macro'),metrics.f1_score(y_test, y_pred, average='macro')])
def _trainModel(algos, descript):
  for i in algos:
    i.fit(X_train, y_train)
    y_pred = i.predict(X_test)
    _insertTable(t, i.__class__.__name__ +" " + descript, y_test, y_pred);

trainSet = pd.read_csv("fashion_train.csv")
testSet = pd.read_csv("fashion_test.csv")
X = trainSet.drop('y', axis=1);
y= trainSet[['y']]
clf=RandomForestClassifier(n_estimators=20)
clf.fit(X,y)
feature_imp = pd.Series(clf.feature_importances_,index=X.columns).sort_values(ascending=False)
# feature_imp
X_train = X[feature_imp.index[0:100]]
X_test = testSet[X_train.columns]
y_train = y
y_test = testSet[['y']]


rdf=RandomForestClassifier(n_estimators=100)
svm_l = svm.SVC(kernel='linear')
model = GaussianNB()
t = PrettyTable(['Algorithms', 'Accurance', 'Precision', 'Recall','F1 measure'])

algos=[svm_l, rdf, model]
_trainModel(algos, " Using selection feature")


X_train = trainSet.drop('y', axis=1)
X_test = testSet.drop('y', axis=1)
y_train = trainSet.y
y_test = testSet.y

_trainModel(algos, " Without using selection feature")
print(t)
```
